### Database


Django framework has a sqlite database by default in root directory, for initializing and migrating tables : 
    
```
>  python manage.py migrate

```









    



After you have created your models on `[AppName]/models.py`, you can append your database changes by creating a migration through the command  : 

```
>  python manage.py makemigrations [MigrationName]

```


    





You can view your database changes as sql statements by executing the command  : 

```
>  python manage.py sqlmigrate [AppName] [MigrationNo]

```


    



### Commandline database manipulation


#### 1 - Inserting into Database 


To insert into Table, simply create the model using the construtor and call on it the save method to be saved :

```python
python_row=Allcourses(coursename="django", insname="XYZ")
python_row.save()

```

#### 2 - Retrieve from Database 


To read from Table, access the model and call `all()` function to retrieve all data :

```python

result=Allcourses.objects.all()


```

you can use `filter()` or `filter()` and pass a value to filter by it
```python

result=Allcourses.objects.filter(id=1)


```
